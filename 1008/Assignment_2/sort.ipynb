```python

def selectionSort(list):
    
    comparisonCounter = 0
    for fillslot in range(len(list) - 1, 0, -1):
        posOfMax = 0
        for location in range(1, fillslot + 1):
            comparisonCounter += 1
            if list[location] > list[posOfMax]:
                posOfMax = location
        temp = list[fillslot]
        list[fillslot] = list[posOfMax]
        list[posOfMax] = temp
    print(f"Number of comparisons: {comparisonCounter}")

def insertSort(list):
    counterComparison = 0
    for index in range(1, len(list)):
        currentVal = list[index]
        pos = index

        while pos > 0 and list[pos - 1] > currentVal:
            counterComparison += 1
            list[pos] = list[pos - 1]
            pos = pos - 1
        list[pos] = currentVal
    print(counterComparison)



def merge(a, b):
    c = []
    sizeA = len(a)
    sizeB = len(b)
    
    indexA = indexB = 0
    while indexA < sizeA and indexB < sizeB:
        if a[indexA] <= b[indexB]:
            c.append(a[indexA])
            indexA += 1
            
        else:
            c.append(b[indexB])
            indexB += 1
            

    for i in range(indexA, sizeA):
        c.append(a[i])

    for i in range(indexB, sizeB):
        c.append(b[i])

    

    return c

def mergeSort(list):
    
    size = len(list)

    if size == 1:
        return list
    
    midIndex = (int) (size / 2)
    firstHalf = list[0:midIndex]
    secondHalf = list[midIndex:size]

    firstHalf = mergeSort(firstHalf)
    secondHalf = mergeSort(secondHalf)


    list = merge(firstHalf, secondHalf)

    return list




# original array
list = [54,26,93,17,77,31,44,55,20]

sortedList = [31, 44, 54, 55, 77, 93]
size6Sorted = [17, 20, 26, 31, 44, 54]
size6Identical = [17,17,17,17,17,17]
size6Unsorted = [54,26,93,17,77,31]
size6Unsorted2 = [77,17,31,44,55,20]

mergeSort(size6Unsorted)
print(size6Unsorted)
print("\n")

```
